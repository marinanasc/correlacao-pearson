```python

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import random
import seaborn as sns
from scipy.stats import pearsonr
```

```python
df = pd.read_csv('excel_ma.csv', sep = ';', decimal = ',', encoding = 'latin-1')
df = df.iloc[:,3:]
```

```python
#eixo X: variáveis de performance
#eixo Y: variáveis de morfologia

vars_morf = ['Comprimento Médio da Fibra  (mm)', 'Coarseness (mg/100m)','Nº de Fibras  (g x 106)', 
             'Finos (%)', 'Vasos   (Nº/grama)','Área Média do Vaso (mm²)', 'Curl  (%)', 'Kink (mm)',
             'Diâmetro da Fibra  (mm)', 'Espessura da Parede (µm)','Diâmetro do Lúmen (mm)', 'Fração Parede (%)']

vars_perform = ['Umidade', 'Gramatura','Viscosidade', 'pH da Polpa para Torres 1 e 2', 'PRESSÃO DO SECADOR',
                'VELOCIDADE TELA MV QCS', 'CONSUMO ESPECÍFICO DE VAPOR','Abertura da Válvula de Vapor p/ Secador', 
                'TEMP. 2 DO SECADOR','Pressao VBP para MC2', 'PRODUÇÃO BRANQUEAMENTO']
```

```python
def grafico_e_corr(x,y):

    fig, ax1 = plt.subplots()

    x = x
    y = y

    print('Coeficiente de Correlação de Pearson:',round(pearsonr(x,y)[0],3))

    ax1.plot(x, y, 'bo')

    plt.show()
```

```python
for i in vars_perform:
    for j in vars_morf:
        print('X:',i, 'Y:',j)
        grafico_e_corr(df[i].values, df[j].values)
```

```python
#mapa de calor

k = 23
corrmat = df.corr()
cols = corrmat.nlargest(k, 'CONSUMO ESPECÍFICO DE VAPOR')['CONSUMO ESPECÍFICO DE VAPOR'].index
f, ax = plt.subplots(figsize=(15, 12))
cm = np.corrcoef(df[cols].values.T)
sns.set(font_scale=0.75)
s = sns.heatmap(cm,  linewidths=0.18,cbar=True, annot=True, square = True, fmt='.2f', 
                annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)
ax.set_ylim(0,23)
s.set_title("Mapa de calor de correlação entre as variáveis", size=12)
plt.show()

plt.savefig('heatmap.png')
```
